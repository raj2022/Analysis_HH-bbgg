```python
import ROOT

ROOT.gROOT.SetBatch(True)

data_file = "output_root/Data_EraE.root"

background_files = [
    "output_root/GGJets.root"
]

tree_name = "DiphotonTree/data_125_13TeV_NOTAG"
branch_name = "HHbbggCandidate_mass"

colors = [
    ROOT.kRed, ROOT.kBlue, ROOT.kGreen, ROOT.kOrange,
    ROOT.kMagenta, ROOT.kCyan, ROOT.kYellow, ROOT.kPink
]

canvas = ROOT.TCanvas("canvas", "Stack Plot", 800, 600)
stack = ROOT.THStack("stack", "Stacked Histograms")
legend = ROOT.TLegend(0.7, 0.7, 0.9, 0.9)

# Load data file
data_chain = ROOT.TChain(tree_name)
data_chain.Add(data_file)

# Create data histogram
data_hist = ROOT.TH1F("data_hist", "", 3500, 0, 3500)
data_chain.Draw("{0}>>data_hist".format(branch_name))
data_hist.SetMarkerStyle(20)
data_hist.SetMarkerColor(ROOT.kBlack)
data_hist.SetLineColor(ROOT.kBlack)
legend.AddEntry(data_hist, "Data", "lep")

# Load background files and create histograms
for i, background_file in enumerate(background_files):
    file = ROOT.TFile.Open(background_file)
    tree = file.Get(tree_name)
    hist = ROOT.TH1F("hist_bg_{0}".format(i), "", 3500, 0, 3500)
    tree.Draw("{0}>>hist_bg_{1}".format(branch_name, i))
    hist.SetFillColor(colors[i])
    hist.SetLineColor(colors[i])
    stack.Add(hist)
    legend.AddEntry(hist, background_file.split('/')[-1].replace('.root', ''), "f")

# Draw stacked histogram
stack.Draw("NOSTACK")

# Draw data histogram on top
data_hist.Draw("SAME P E1")

# Draw legend
legend.Draw()

# Save canvas as a PDF file
canvas.SaveAs("stack_plot.pdf")
```

```python
import uproot
import numpy as np
import matplotlib.pyplot as plt

# List of background files
background_files = [
    "output_root/GGJets.root",
    "output_root/GJetPt20To40.root",
    "output_root/GJetPt40.root",
    "output_root/GluGluHToGG.root",
    "output_root/GluGluToHH.root",
    "output_root/VBFHToGG.root",
    "output_root/VHToGG.root"
]

# Variable to plot (pt)
variable = "jet1_pt"

# Create an empty list to store histograms
histograms = []

# Create a figure and axis for plotting
fig, ax = plt.subplots()

# Loop over each background file
for file_name in background_files:
    # Open the ROOT file and access the specified TTree
    with uproot.open(file_name) as file:
        tree = file["DiphotonTree/data_125_13TeV_NOTAG"]

        # Load the data from the TTree
        data = tree.array(variable)

        # Create a histogram for this background sample
        hist, bins = np.histogram(data, bins=100, range=(0, 100))

        # Add the histogram to the list
        histograms.append(hist)

# Stack the histograms
stacked_histogram = np.vstack(histograms)

# Plot the stacked histogram
ax.hist(bins[:-1], bins, weights=stacked_histogram, stacked=True, label=background_files)

# Customize the plot
ax.set_xlabel("pT [GeV]")
ax.set_ylabel("Events")
ax.set_title("Stacked Histogram of Backgrounds (pT)")
ax.legend()

# Show the plot
plt.show()
```

```python
import ROOT

# Set the batch mode
ROOT.gROOT.SetBatch()
ROOT.gStyle.SetOptStat(0)

# Define the file paths
file_paths = [
    "output_root/GGJets.root",
    "output_root/GJetPt20To40.root",
    "output_root/GJetPt40.root",
    "output_root/GluGluHToGG.root",
    "output_root/GluGluToHH.root",
    "output_root/VBFHToGG.root",
    "output_root/VHToGG.root",
    "output_root/ttHToGG.root"
]

# Define the variable to plot (e.g., pt)
variable = "eta"

# Create a canvas to draw the histograms
canvas = ROOT.TCanvas('canvas', "", 600, 600)

# Create THStack for the stacked histogram
stack = ROOT.THStack("", "")

# Define colors for each background
colors = [ROOT.kRed, ROOT.kBlue, ROOT.kGreen, ROOT.kYellow, ROOT.kCyan, ROOT.kMagenta, ROOT.kOrange, ROOT.kPink]

# Create a legend
legend = ROOT.TLegend(0.7, 0.7, 0.9, 0.9)

# Open each file, create a histogram, and draw it on the canvas
for i, file_path in enumerate(file_paths):
    # Open the ROOT file
    file = ROOT.TFile(file_path, "READ")

    # Access the TTree containing the data
    tree = file.Get("DiphotonTree/data_125_13TeV_NOTAG")

    # Create a histogram from the TTree and the variable
    hist = ROOT.TH1F("", "", 100, 0, 100)  # Adjust the binning as needed
    tree.Project(hist.GetName(), variable)

    # Set histogram attributes
    hist.SetFillColor(colors[i % len(colors)])  # Set a different color for each file
    hist.SetLineColor(ROOT.kBlack)  # Set the line color
    hist.SetLineWidth(2)  # Set the line width

    # Add histogram to the stack
    stack.Add(hist)

    # Draw the histogram
    if i == 0:
        hist.Draw("hist")
    else:
        hist.Draw("hist same")

    # Add entry to the legend
    legend.AddEntry(hist, file_path, "f")

# Set logarithmic scale for the y-axis
canvas.SetLogy()

# Draw the legend
legend.Draw()

# Draw the canvas
canvas.Draw()
```
